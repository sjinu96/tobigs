# feature 별 특성

> **'datetime' 열의 datatype을 'datetime64' 로 변경(이전 : str)**

```python
dat['datetime']=dat['datetime'].astype('datetime64')
```

> **year, month, day, time feature 추가하기**

```python
dat['year']=dat['datetime'].apply(lambda x : x.year)
dat['month']=dat['datetime'].apply(lambda x : x.month)
dat['day']=dat['datetime'].apply(lambda x : x.day)
dat['hours']=dat['datetime'].apply(lambda x : x.hour)
```

```python
dat.groupby('year').sum()
```

```python
dat
```

```python
figure, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2)
figure.set_size_inches(18,8)

sns.barplot(data=dat.groupby('year').sum().reset_index(),x='year',y="count", ax=ax1)


ax1.set(ylabel='Count',title="연도별 대여량")
ax2.set(xlabel='month',title="월별 대여량")
ax3.set(xlabel='day', title="일별 대여량")
ax4.set(xlabel='hour', title="시간별 대여량")
```
